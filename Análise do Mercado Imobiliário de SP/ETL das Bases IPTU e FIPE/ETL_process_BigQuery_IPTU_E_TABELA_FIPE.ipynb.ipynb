## O que é o processo ETL?

Extrair, transformar e carregar (ETL) é o processo que as organizações orientadas a dados usam para coletar dados de várias fontes e reuni-los para dar suporte à descoberta, à geração de relatórios, à análise e à tomada de decisões.

As origens de dados podem ser muito diversas em tipo, formato, volume e confiabilidade, de modo que os dados precisam ser processados para serem úteis quando reunidos. Os armazenamentos de dados de destino podem ser bancos de dados, data warehouses ou data lakes, dependendo das metas e da implementação técnica.



    O processo de ETL nas bases de dados 'IPTU' e 'TABELA FIPE' foi realizada no Big Query;

# IPTU São Paulo 2022

O dataset contém os registros em cartório do contribuinte em relação ao IPTU no ano de 2022, na cidade de São Paulo.

    Cada linha da tabela corresponde a uma informação sobre o imóvel.

O IPTU (Imposto Predial e Territorial Urbano) é um imposto municipal cobrado sobre propriedades urbanas, sejam elas terrenos, edificações ou uma combinação de ambos.

É uma das principais fontes de receita para os municípios e é utilizado para financiar serviços públicos, como infraestrutura urbana, saúde, educação, segurança e outros programas municipais.


- [Fonte de dados](https://www.kaggle.com/datasets/andasampa/iptu-sao-paulo)
