Ao investigar meticulosamente variáveis específicas, é possível obter insights profundos que transcendem os números brutos, proporcionando uma compreensão mais completa das nuances que orientam as decisões de compra, venda e investimento.

Iniciando as análises das variáveis que impactam o mercado imobiliário, baseada em dados estatísticos do município de São Paulo, nosso objetivo é mergulhar nas complexas interações que moldam esse setor dinâmico.

## Área do terreno (m²) *versus* Área construída (m²)

image.png

* Ao comparar essas duas métricas, podemos entender como a proporção entre elas impacta o valor e a utilização das propriedades.

      No geral, não há proporção entre área de terreno e área construída.

* Na década de 70, mais especificamente 1974, havia uma proporcionalidade entre estas variáveis. A partir de então, houve uma surpreendente alta na área construída.

      Boom da verticalização nas décadas de 70 e 80;
      Transformação da paisagem urbana;
      Construção de 1,2 milhão de apartamentos ao longo de seis décadas;
      A década de 70 foi marcante nesse processo.

* Ao longo do período entre 2013 e 2021, houve uma tendência de queda na área construída. Tal pode ser atribuída a uma combinação de fatores econômicos, políticos e sociais que impactaram o mercado imobiliário durante esse período.
      
      O Brasil em crise econômica e com alta taxa de desemprego;
      Diminuição da capacidade das pessoas de adquirirem imóveis;
      Redução na oferta de crédito (dificuldade de financiamento);
      Impacto doreto sobre a demanda no mercado imobiliário;

Esta análise sugere que, ao longo das décadas, houve momentos de otimização da utilização do espaço construído e do terreno disponível.

  - Tendências decrescentes subsequentes na área construída podem sugerir preferência por propriedades mais compactas ou uma adaptação às demandas do mercado.
  - Aumento expressivo na área do terreno em 2012 pode estar relacionado a mudanças regulatórias ou oportunidades de expansão nesse período.




    Atenção à discrepância nas métricas de construção e terreno:
      - Podem refletir mudanças socioeconômicas, regulatórias ou de demanda;
      - Impactam o mercado imobiliário;
      - Influenciam estratégias de investimento e desenvolvimento.

## Áreas construídas *versus* quantidade de pavimentos

Relacionar a área construída com quantidade de pavimentos, essencial para avaliar o valor do imóvel.

image.png

* A maioria dos imóveis possui até 02 pavimentos, o que poderia indicar uma prevalência de casas ou prédios de pequeno porte.

* Tratando-se de padrões de urbanização, esse tipo de imóvel pode estar alocado em áreas mais suburbanas.

      Menor densidade populacional nessas regiões;
      Precedência por locais mais espaçosos/menos densos;
      98% destes imóveis são do tipo residencial.

* Presença significativa de imóveis com mais de 16 a 21 pavimentos sugere uma concentração de edifícios relacionados a áreas urbanas mais densamente povoadas ou a regiões mais desenvolvidas economicamente.

      Desenvolvimento vertical em áreas com crescimento econômico acelerado;
	  Indicação de maior crescimento econômico e oportunidades de emprego;
      Investimentos na verticalidade para receber a demanda de negócios.

* Tratando-se de imóveis de 3 a 5 pavimentos (intermediário), pode-se rematá-los como locais de construções multifamiliares.

      Acomodar um quantitativo expressivo de famílias;
      Adaptação de habitações mais compactas.


Seria interessante explorar ainda mais esses insights considerando fatores como zonas geográficas específicas em São Paulo.

Todavia, as bases analisadas neste estudo não continham elementos suficientes para se trabalhar com esse viés.

# Análise SWOT

A análise SWOT é fundamental para análise de dados, pois traz um arcabouço que estabelece informações e identifica padrões e relações nos dados.

image.png

* Forças


      - Diversidade e a dinâmica do mercado (oportunidades);
      - Adaptação às demandas dos compradores e dos padrões de urbanização;
      - Investidores em busca de oportunidades de valorização.

* Fraquezas


      - Mais rigor na fiscalização para prevenir evasão fiscal;
      - Condições macroeconômicas podem impactar negativamente esse nicho;
      - Períodos de quedas nas transações (desafios para o mercado imobiliário).

*	Oportunidades


        - Expansão contínua, com uma base de demanda sólida;
        - Espaços mais eficientes/tópico ecológico responsável;
        - Regularição e incentivos à formalização de transações.

* Ameaças


      - Garantir a transparência e a equidade no mercado;
      - Mudanças em estilos de vida influenciam diretamente a demanda;
      - Exposição do mercado a riscos financeiros, especialmente se ocorrerem altos índices de inadimplência.




# Conclusões

Ao explorar a vinculação das variáveis que impactam o mercado imobiliário, foi demonstrado como fatores históricos, econômicos e sociais adaptaram a evolução desse nicho tão dinâmico ao longo do tempo.

O objetivo de compreender as complexidades do mercado e identificar os principais fatores que influenciam as decisões dos compradores foram alcançadas com sucesso.

## Acessibilidade e poder de compra

* A análise das transações de compra e venda em relação aos valores declarados e venais destaca a importância de avaliações precisas e fiscalização rigorosa para garantir a equidade fiscal e a integridade do mercado.

* A movimentação entre taxas de juros, medidas governamentais durante a pandemia ressaltou a dependência entre condições econômicas e atividade do mercado imobiliário.

## Fatores influentes na decisão do Comprador

* A predominância de imóveis de até dois pavimentos, juntamente com a observação das transações de alienação fiduciária e execuções extrajudiciais, destaca a influência das condições financeiras individuais e contextos externos nas decisões dos compradores e nas dinâmicas do mercado imobiliário.

## Perfis dos imóveis

* Foi possível traçar perfis detalhados que abrangem desde a preferência por áreas suburbanas até a concentração de edifícios em regiões urbanas mais desenvolvidas economicamente.

* Esses perfis fornecem insights valiosos para o planejamento urbano eficaz e o direcionamento estratégico de investimentos no mercado imobiliário.

## Tendências e padrões

* Integrar à análise indicadores gerais de registro imobiliário enriqueceu o processo, revelando padrões de recuperação após um declínio no mercado e a preferência crescente por transações de compra e venda.

* Essas tendências sugerem uma interação complexa entre fatores econômicos, políticos e sociais que influenciam as atividades do mercado imobiliário e indicam oportunidades para o desenvolvimento de estratégias futuras.

      A análise proporcionou uma visão esclarecedora do mercado imobiliário;
      
      Foram descritos fatores integrados que configuram:
      - Comportamento dos compradores;
      - Tendências de transações;
      - Dinâmica geral do setor.
